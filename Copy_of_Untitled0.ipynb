```python
var = np.arange(0,10)
```

```python
var1 = var ** 2
```

```python
print(var1)
```

```python
plt.plot(var, 'red', label = 'price of oil')
plt.plot(var1, 'blue', label = 'price of gold')
plt.xlabel('time')
plt.ylabel('price')
plt.title('time vs price')
plt.legend(title = 'variables')
plt.show()
```
